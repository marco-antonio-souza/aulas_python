Verificar número positivo
Peça ao usuário um número e verifique: Se o número for maior que 0, mostre "Número positivo". (Use apenas if).

Verificar maior de idade
Peça a idade do usuário. Mostre "Maior de idade" se a idade for maior ou igual a 18. (Use apenas if).

Número maior que 10
Peça um número ao usuário. Se o número for maior que 10, mostre "Número maior que 10". (Use apenas if).


```python
numero = int(input("Digite um numero"))

if numero >= 0: 
    print("O numero é positivo")
else:
    print("O numero é negativo")
```

```python
numero = int(input("Digite a sua idade"))

if numero >= 18: 
    print("Voce é maior de idade")
else:
    print("Voce é menor de idade")
```

```python
numero = int(input("Digite um numero"))

if numero >= 10: 
    print("O numero é maior que 10")
else:
    print("O numero é menor que 10")
```

Comparar dois números
Peça dois números ao usuário. Mostre qual é maior. Exemplo: "O maior número é 25". (Use if e else).

```python
numero = int(input("Digite um numero"))

if numero % 2: 
    print("O numero é par")
else:
    print("O numero é impar")
```

```python
nota = int(input("Digite uma nota"))

if nota >= 7: 
    print("Aluno aprovado")
else:
    print("Aluno reprovado")
```

```python
numero = int(input("Digite um numero"))

if numero != 25:
    print("O numero é maior que 25")
else:
    print("O numero é menor que 25")
```

Crie um programa que peça um número e mostre:

▶ Número positivo

▶ Número negativo

▶ Número igual a zero

Dica: Use as estruturas if, elif e else para resolver esse problema.

```python
numero = int(input("Digite um numero"))

if numero >= 0:
    print("Numero positivo")
elif numero <= 0:
    print("Numero negativo")
else:
    print("Numero igual a zero")
```
